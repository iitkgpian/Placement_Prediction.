Calculate P(X=x|Y=y) Categorically

```python
def calculate_categorical(data, feat_name, feat_value, Y, label):
  feat = list(data.columns)
  data = data[data[Y]==label]
  p_x_given_y = len(data[data[feat_name]==feat_value]) / len(data)
  return p_x_given_y
```

Calculate P(Y = y) for all possible Y

```python
def calculate_prior(data, Y):
  classes = sorted(list(data[Y].unique()))
  prior = []

  for i in classes:
    prior.append(len(data[data[Y]==i])/len(data))
  return prior
```

Calculate P(X=x1|Y=y)P(X=x2|Y=y)....P(X=xn|Y=y) * P(Y=y) for all Y and find the maximum-

```python
def naive_bayes_categorical(data, X, Y):
  # features name
  features = list(data.columns)[:-1]

  prior = calculate_prior(data, Y) #list containing probability of placed and non placed data

  Y_pred = []

  #loop for every data sample
  for x in X:
    labels = sorted(list(data[Y].unique()))
    likelyhood = [1]*len(labels)
    for j in range(len(labels)):
      for i in range(len(features)):
        likelyhood[j] *= calculate_categorical(data, features[i], x[i], Y, labels[j])

    #calculate posterior probability numerator only
    post_prob = [1]*len(labels)
    for j in range(len(labels)):
      post_prob[j] = likelyhood[j] * prior[j]

    Y_pred.append(np.argmax(post_prob))

  return np.array(Y_pred)



```

test categorical model

For the dataset dataset2

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(dataset2, test_size = .2, random_state = 42)

X_test = test.iloc[:,:-1].values
Y_test = test.iloc[:,-1].values
Y_pred = naive_bayes_categorical(train, X=X_test, Y="Paced or Not Placed")

from sklearn.metrics import confusion_matrix, f1_score
print(confusion_matrix(Y_test, Y_pred))
f1_score(Y_test, Y_pred)
```

Inbuild  Categorical Naive Bayes

```python
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import confusion_matrix, f1_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
X5_data = dataset5.iloc[:,:-1].values
Y5_data = dataset5.iloc[:,-1].values

X5_train, X5_test, Y5_train, Y5_test = train_test_split(X5_data, Y5_data, test_size=0.2, random_state=42)
```

```python
model = CategoricalNB()
model.fit(X5_train, Y5_train)
```

```python
Y5_pred = model.predict(X5_test)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(Y5_test, Y5_pred))
print("\nf1_score:")
print(f1_score(Y5_test, Y5_pred))
```
